## Components of dimensionality reduction
- Feature selection - select a subset
- Feature extraction - build a new feature from original feature set

```python
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
X = iris.data
Y = iris.target
print(X)
print(Y)
```

```python
pca = PCA(n_components=2)
X_R = pca.fit(X).transform(X)
print(X_R)
```

```python
lda = LDA(n_components=2)
X_L = lda.fit(X,Y).transform(X)
print(X_L)
```

```python
from sklearn.preprocessing import MinMaxScaler

data = [[3,4], [5,6], [1,2], [4,9]]
scaler = MinMaxScaler()
scaler.fit(data).transform(data)
```

```python
scaler.fit(X_R).transform(X_R)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data = ["america","india","british","china","india"]
le.fit(data)
encoded = le.transform(data)
print(encoded)
```

```python
le.inverse_transform(encoded)
```
